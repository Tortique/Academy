Nettoyage des données :
- Le nombre d'indicateurs peut être réduits afin de correspondre à la population ciblés par la société : Population de +15 ans, étudiants lycée et université, accès internet.
- On peut constater, avec les valeurs manquantes dans Data, que les années possédant le plus de données sont entre 2010 et 2015, inclus. 
- On peut aussi réduire les pays cibles en se basant sur le PIB, PIB / habitant et le niveau de revenu, et aussi supprimer les regroupements de pays.

Ce qui peut déjà permettre de réduire le nombre de cellules à traiter pour les prochaines étapes
On garde donc ces valeurs dans des variables min et max, afin de pouvoir les modifier plus facilement

* Tri des indicateurs

D'abord, on récupère les données sur la population

```python
population = series.loc[series['Topic'].str.contains('Population') & series['Indicator Name'].str.contains('15-64') & series['Indicator Name'].str.contains('total')]
population
```

Ensuite, le niveau d'étude

```python
indicators = ['SE.SEC.ENRR','SE.TER.ENRR']
students = series.loc[series['Series Code'].isin(indicators)]
students
```

Ensuite l'accès à internet

```python
internet_access = series.loc[series['Long definition'].str.contains('Internet')]
internet_access
```

On cherche des indicateurs sur le niveau de revenu des pays (PPA : 'parité de pouvoir d'achat' ou PPP en anglais)

```python
countries_indicators = series.loc[series['Indicator Name'].str.contains('GNI per capita, PPP')]
countries_indicators
```

```python
indicators = pd.concat([population,students, internet_access,countries_indicators], ignore_index=True)
indicators
```

```python
print(f"On se retrouve donc avec {indicators.index.size} indicateurs potentiels, pouvant correspondre à notre population")
```

On cherche tous les regroupements de pays, afin de les supprimer

```python
group = countries.loc[countries['Special Notes'].str.contains('aggregate', na=False)]
```

```python
data = data[~data['Country Code'].isin(group['Country Code'].tolist())]
data
```

* On réduit le nombre de colonnes et de lignes dans data en réduisant le nombre de dates et en ne prenant que les indicateurs potentiels

```python
date_min = 2010
date_max = 2015
```

```python
data = data.loc[data['Indicator Code'].isin(indicators['Series Code'].tolist())]
data
```

```python
from datetime import date

for col in data.select_dtypes(include=np.number).columns:
    if not date_min <= date(int(col),1,1).year <= date_max:
        del data[col]
data
```

```python
msno.bar(data)
```

```python
data_melt = pd.melt(data, id_vars=['Country Name','Country Code','Indicator Name','Indicator Code'], value_vars=data.select_dtypes(include=np.number).columns.tolist())
data_melt
```

```python
data_pivot = pd.pivot_table(data_melt, values='value',columns='Indicator Code', index=['Country Code','Country Name'],aggfunc=np.mean)
data_pivot
```

```python
msno.matrix(data_pivot)
```

```python
to_drop = data_pivot[data_pivot.isna().any(axis=1)].index.get_level_values(0).unique().tolist()
data_pivot= data_pivot[~data_pivot.index.get_level_values(0).isin(to_drop)]
data_pivot.reset_index()
data_pivot
```

```python
data_pivot = data_pivot.rename(columns = {'IT.NET.USER.P2' : 'NET','NY.GNP.PCAP.PP.CD': 'GDP', 'SE.SEC.ENRR' : 'SEC', 'SE.TER.ENRR' : 'TER',
                                          'SP.POP.1564.TO' : 'POP TOT', 'SP.POP.1564.TO.ZS' : 'POP %'})
```

```python
data_pivot['Score'] = (((data_pivot['Top NET']*6)+(data_pivot['Top TER']*5)+(data_pivot['Top SEC']*4)+(data_pivot['Top GDP']*3)+(data_pivot['Top 1564']*2)+(data_pivot['Top POP']*1))/21).rank()
best_countries = data_pivot.sort_values(by='Score').reset_index()
best_countries
```

```python
import requests

geojson_data = requests.get(
    "https://raw.githubusercontent.com/python-visualization/folium-example-data/main/world_countries.json"
).json()

map = folium.Map(location=(30, 10), zoom_start=3, tiles="cartodb positron")
folium.Choropleth(
    geo_data=geojson_data,
    data=best_countries,
    columns=['Country Name', 'Score'],
    key_on='feature.properties.name',
    bins=[0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, data_pivot['Score'].max()],
    fill_color="RdYlGn_r",
    fill_opacity=0.8,
    line_opacity=0.3,
    nan_fill_color="white",
    legend_name="Pays Potentiels",
).add_to(map)

map
```

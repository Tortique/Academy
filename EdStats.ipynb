Conclusion sur les données :
    - EdStatsCountry comprends les données des pays par code unique, avec leurs noms, leurs régions et leurs groupes de revenus.
    - EdStatsCountry-Series reprends les code uniques des pays ainsi que les indicateurs avec une description des sources des données.

    - EdStatsSeries comprends le code unique des indicateurs, leurs sujets et le nom
    - EdStatsData comprends les résultats des indicateurs par code pays et année (colonnes)

Nettoyage des données :
    - On peut constater, avec les valeurs manquantes dans Data, que les années possédant le plus de données sont entre 1990 et 2015, inclus. 
    - Ce qui peut déjà permettre de réduire le nombre de cellules à traiter pour les prochaines étapes
On garde donc ces valeurs dans des variables min et max, afin de pouvoir les modifier plus facilement

```python
date_min = 2010
date_max = 2015
```

```python
from datetime import date

for col in data.select_dtypes(include=np.number).columns:
    if not date_min <= date(int(col),1,1).year <= date_max:
        del data[col]
data
```

```python
msno.bar(data)
```

```python
data_melt = pd.melt(data, id_vars=['Country Name','Country Code','Indicator Name','Indicator Code'], value_vars=data.select_dtypes(include=np.number).columns.tolist())
data_melt
```

```python
count = data_melt.groupby(['Indicator Code'])['value'].count().reset_index()
count = count.drop(count[count.value < 500].index)
count
```
